```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinson_data=pd.read_csv('/content/parkinsons.csv')
```

```python
parkinson_data.head()
```

```python
parkinson_data.shape
```

```python
parkinson_data.info()
```

```python
parkinson_data.isnull().sum()
```

```python
parkinson_data.describe()
```

```python
#to see how many ppl have parkinsons disease and how many do not in the dataset
parkinson_data['status'].value_counts()
```

seperating features and target

```python
X=parkinson_data.drop(columns=['name','status'], axis=1)
Y=parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

splitting data into testing and training data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

standardising the data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model training

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

Evaluating the model(accuracy score)

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("accuracy of training data= ",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print("accuracy of test data= ",test_data_accuracy)
```
